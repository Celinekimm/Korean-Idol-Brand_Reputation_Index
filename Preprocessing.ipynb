```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://bit.ly/ds-korean-idol')
```

```python
df.info()
```

```python
# select_dtypes
df.select_dtypes(include='object')
```

```python
df.select_dtypes(exclude='object')
```

```python
# One-hot-encoding
blood_map = {
    'A':0,
    'B':1,
    'AB':2,
    'O':3,
}
```

```python
df['혈액형_code'] = df['혈액형'].map(blood_map)
```
